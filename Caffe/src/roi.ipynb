```python
import tables
import numpy
import h5py
import os
```

```python
RAW_RESPONSES_PATH = "../data/EstimatedResponses.mat"
STIMULI_TRAIN_PATH = "../data/StimuliTrain.h5"
STIMULI_VALIDATION_PATH = "../data/StimuliValidation.h5"

RAW_RESPONSES = tables.openFile(RAW_RESPONSES_PATH)
RAW_STIMULI_TRAIN = tables.openFile(STIMULI_TRAIN_PATH)
RAW_STIMULI_VALIDATION = tables.openFile(STIMULI_VALIDATION_PATH)
```

```python
RESPONSE_TRAIN_ROI_S1 = []
RESPONSE_TRAIN_ROI_S2 = []

RESPONSE_TRAIN_S1 = RAW_RESPONSES.getNode('/dataTrnS1')[:]
RESPONSE_TRAIN_S2 = RAW_RESPONSES.getNode('/dataTrnS2')[:]

ROI_S1 = RAW_RESPONSES.getNode('/roiS1')[:].flatten()
ROI_S2 = RAW_RESPONSES.getNode('/roiS2')[:].flatten()

# #build roi dictionaries
for response in RESPONSE_TRAIN_S1:
    idx = 0
    roi_vals = {}
    response_roi_avg = []
    
    for roi in ROI_S1:
        if roi not in roi_vals:
            roi_vals[roi] = []
            
        if not numpy.isnan(response[idx]): 
            roi_vals[roi].append(response[idx])
            
        idx += 1
    
    for i in range(0, 8):
        val = sum(roi_vals[i]) / len(roi_vals[i])
        response_roi_avg.append(val)
        
    RESPONSE_TRAIN_ROI_S1.append(response_roi_avg)
    
for response in RESPONSE_TRAIN_S2:
    idx = 0
    roi_vals = {}
    response_roi_avg = []
    
    for roi in ROI_S2:
        if roi not in roi_vals:
            roi_vals[roi] = []
            
        if not numpy.isnan(response[idx]): 
            roi_vals[roi].append(response[idx])
            
        idx += 1
    
    for i in range(0, 8):
        val = sum(roi_vals[i]) / len(roi_vals[i])
        response_roi_avg.append(val)
        
    RESPONSE_TRAIN_ROI_S2.append(response_roi_avg)
    
RESPONSE_TRAIN_ROI_S1 = numpy.asarray(RESPONSE_TRAIN_ROI_S1) 
RESPONSE_TRAIN_ROI_S2 = numpy.asarray(RESPONSE_TRAIN_ROI_S2)
```

```python
RESPONSE_VALIDATION_ROI_S1 = []
RESPONSE_VALIDATION_ROI_S2 = []

RESPONSE_VALIDATION_S1 = RAW_RESPONSES.getNode('/dataValS1')[:]
RESPONSE_VALIDATION_S2 = RAW_RESPONSES.getNode('/dataValS2')[:]

# #build roi dictionaries
for response in RESPONSE_VALIDATION_S1:
    idx = 0
    roi_vals = {}
    response_roi_avg = []
    
    for roi in ROI_S1:
        if roi not in roi_vals:
            roi_vals[roi] = []
            
        if not numpy.isnan(response[idx]): 
            roi_vals[roi].append(response[idx])
            
        idx += 1
    
    for i in range(0, 8):
        val = sum(roi_vals[i]) / len(roi_vals[i])
        response_roi_avg.append(val)
        
    RESPONSE_VALIDATION_ROI_S1.append(response_roi_avg)
    
for response in RESPONSE_VALIDATION_S2:
    idx = 0
    roi_vals = {}
    response_roi_avg = []
    
    for roi in ROI_S2:
        if roi not in roi_vals:
            roi_vals[roi] = []
            
        if not numpy.isnan(response[idx]): 
            roi_vals[roi].append(response[idx])
            
        idx += 1
    
    for i in range(0, 8):
        val = sum(roi_vals[i]) / len(roi_vals[i])
        response_roi_avg.append(val)
        
    RESPONSE_VALIDATION_ROI_S2.append(response_roi_avg)
    
RESPONSE_VALIDATION_ROI_S1 = numpy.asarray(RESPONSE_VALIDATION_ROI_S1)
RESPONSE_VALIDATION_ROI_S2 = numpy.asarray(RESPONSE_VALIDATION_ROI_S2)
```

```python
STIMULI_TRAIN = RAW_STIMULI_TRAIN.getNode('/C:\data')[:]
STIMULI_VALIDATION = RAW_STIMULI_VALIDATION.getNode('/C:\data')[:]

STIMULI_TRAIN = numpy.transpose(STIMULI_TRAIN)
STIMULI_VALIDATION = numpy.transpose(STIMULI_VALIDATION)
```

```python
dirname = os.path.abspath('../data/caffe/')
if not os.path.exists(dirname):
    os.makedirs(dirname)

train_filename = os.path.join(dirname, 'train.h5')
test_filename = os.path.join(dirname, 'test.h5')

with h5py.File(train_filename, 'w') as f:
    f['data'] = STIMULI_TRAIN
    f['label'] = RESPONSE_TRAIN_ROI_S1
with open(os.path.join(dirname, 'train.txt'), 'w') as f:
    f.write(train_filename + '\n')

    
with h5py.File(test_filename, 'w') as f:
    f['data'] = STIMULI_VALIDATION
    f['label'] = RESPONSE_VALIDATION_ROI_S1
with open(os.path.join(dirname, 'test.txt'), 'w') as f:
    f.write(test_filename + '\n')
    
# # HDF5 is pretty efficient, but can be further compressed.
# comp_kwargs = {'compression': 'gzip', 'compression_opts': 1}
# with h5py.File(test_filename, 'w') as f:
#     f.create_dataset('data', data=Xt, **comp_kwargs)
#     f.create_dataset('label', data=yt.astype(np.float32), **comp_kwargs)
# with open(os.path.join(dirname, 'test.txt'), 'w') as f:
#     f.write(test_filename + '\n')
```
